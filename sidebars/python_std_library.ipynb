## Itertools and Functools

### Itertools

```python
import itertools
```

```python
help(itertools)
```

```python
my_string = "hello there everybody"
```

```python
cycle = itertools.cycle(my_string)
```

```python
count = 1
while count <=10:
    print(cycle.__next__())
    count += 1
```

## Functools

```python
import functools
```

```python
help(functools)
```

```python
num_list = [1,2,4,5,6,7,9]
```

```python
functools.reduce(lambda x, y: x * y, num_list)
```

### Answers

```python
my_string = "hello there everybody"
cycle = itertools.cycle(my_string)
count = 1
while count <=10:
    print(cycle.__next__())
    count += 1


num_list = [1,2,4,5,6,7,9]
functools.reduce(lambda x, y: x * y, num_list)
```
