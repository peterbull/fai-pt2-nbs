```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(15, 5))
ax1 = plt.axes(131, projection='3d')
ax2 = plt.axes(132, projection='3d')
ax3 = plt.axes(133, projection='3d')

x = np.outer(np.linspace(-3, 3, 30), np.ones(30))
y = x.copy().T
z = np.sin(np.sqrt(x**2 + y**2))

surf1 = ax1.plot_surface(x, y, z, cmap="inferno")
surf2 = ax2.scatter3D(x, y, z, color='red')
surf3 = ax3.plot_surface(x, y, z, cmap='viridis')

ax3.view_init(elev=50, azim=35)

for ax in [ax1, ax2, ax3]:
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.sin(np.sqrt(x**2 + y**2))

surf = ax.plot_surface(x, y, z, cmap='viridis')

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.sin(np.sqrt(x**2 + y**2))

# 3D surface plot
surf = ax1.plot_surface(x, y, z, cmap='viridis')

# Contour plot
contour = ax2.contourf(x, y, z, cmap='viridis')

# Adding color bar to each subplot
fig.colorbar(surf, ax=ax1, shrink=0.6)
fig.colorbar(contour, ax=ax2, shrink=0.6)

plt.show()
```
