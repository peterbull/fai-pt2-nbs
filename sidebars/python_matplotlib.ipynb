```python
import matplotlib.pyplot as plt
```

```python
x = list(range(10))
```

```python
y = x
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color="red", linestyle='dashed')
plt.show()
```

```python
plt.plot(x,y, color="red", linestyle='dashed')
plt.title("Test Graph")
plt.xlabel("widgets")
plt.ylabel("thunderstorms")
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='dashed', label='widgets')
plt.plot(x, [i**2 for i in x], color='blue', linestyle='dotted', label='widgets squared')
plt.title('widget graph')
plt.xlabel('widget 1')
plt.ylabel('widget 2')
```
