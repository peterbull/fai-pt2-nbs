```python
import numpy as np

months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
sales_2022 = np.random.randint(20, 50, 12)
sales_2023 = np.random.randint(30, 60, 12)
comb_sales = np.concatenate([sales_2022, sales_2023])
```

```python
plt.figure(figsize=(20,10))
plt.plot(months, sales_2022, 'g--s', label="2022 Sales")
plt.plot(months, sales_2023, 'r-o', label="2023 Sales")
plt.title('Products Sold By Month')
plt.xlabel('Month')
plt.ylabel('Amount of Products Sold')
plt.legend()
plt.grid(True)
plt.xlim(months[0], months[-1])
plt.ylim(comb_sales[np.argmin(comb_sales)], comb_sales[np.argmax(comb_sales)])
plt.show
```
